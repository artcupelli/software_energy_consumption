```python
import pandas as pd
import pymongo
import plotly.express as px
```

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient['db_sensor']
```

```python
data = pd.DataFrame(list(mydb['report_0'].find({})))
```

```python
data.shape
```

```python
data.reset_index(inplace=True)

for index, row in data.iterrows():
    data.loc[index, "power"] = data.iloc[index].groups['rapl']['0']['7']['RAPL_ENERGY_PKG'] / 1000000
```

```python
px.line(data[data.timestamp > "2024-06-17 00:42:28.049" ], 'timestamp', 'power')
```
